## Q1. 매년 5월달의 open가격의 평균값을 데이터 프레임으로 표현하라

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/nflx/NFLX.csv")

#Date 컬럼을 datatime으로 변경

df["year"] = pd.to_datetime(df["Date"]).dt.year
df["month"] = pd.to_datetime(df["Date"]).dt.month

df_5 = df.loc[df["month"] == 5]
df_5.groupby("year").mean()["Open"].to_frame()
```

```python
#Answer

import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/nflx/NFLX.csv")

df["Date"] = pd.to_datetime(df["Date"])

target = df.groupby(df["Date"].dt.strftime("%Y-%m")).mean()["Open"]

target.loc[target.index.str.contains("-05")].to_frame()
```
