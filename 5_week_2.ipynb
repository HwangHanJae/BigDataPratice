##Q1. bedrooms의 빈도가 가장 높은 값을 가지는 데이터들의 price의 상위 10%와 하위 10% 값의 차이를 구하여라

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/kingcountyprice/train.csv")
#bedrooms의 빈도가 가장 높은 값 찾기
count_high = df.groupby("bedrooms").size().sort_values(ascending=False).index[0]
#빈도가 가장 높은 값은 3

#price의 상위 10% 값
price_high = df.loc[df["bedrooms"] == count_high, "price"].quantile(0.9)

#pricedml 하위 10% 값
price_row = df.loc[df["bedrooms"] == count_high, "price"].quantile(0.1)

print(price_high - price_row)
```

```python
#Answer

df.loc[df["bedrooms"] == df["bedrooms"].value_counts().index[0], "price"].quantile(0.9) \
- df.loc[df["bedrooms"] == df["bedrooms"].value_counts().index[0], "price"].quantile(0.1)

```

#작업형2

데이터 출처 :https://www.kaggle.com/harlfoxem/housesalesprediction (참고, 데이터 수정)

데이터 설명 : 킹카운티 주거지 가격 예측문제

​평가지표 : r2 score

trainData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/kingcountyprice/train.csv

testData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/kingcountyprice/test.csv

```python
trainData = "https://raw.githubusercontent.com/Datamanim/datarepo/main/kingcountyprice/train.csv"
testData = "https://raw.githubusercontent.com/Datamanim/datarepo/main/kingcountyprice/test.csv"

import pandas as pd

train = pd.read_csv(trainData)
test = pd.read_csv(testData)

train
```

```python
#불필요한 컬럼 제거
drop_col = ["date", "id", "zipcode"]

train = train.drop(drop_col, axis=1)
test = test.drop(drop_col, axis=1)

#타겟데이터 분리
X = train.drop("price", axis=1)
y = train["price"]
```

```python
#훈련데이터, 검증데이터
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.33, random_state=42)

#모델링
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()

rf.fit(X_train, y_train)

y_pred = rf.predict(X_val)

#평가
from sklearn.metrics import r2_score

print(r2_score(y_val, y_pred))
```

```python
#제출
#submission 파일 생성

import numpy as np
sub = {"price" : np.zeros(test.shape[0])}
submission = pd.DataFrame(sub)

pred = rf.predict(test)
submission["price"] = pred
submission
```

```python
submission.to_csv("수험번호.csv", index=False)
```
