##Q1. Arrival Delay in Minutes 컬럼이 결측치인 데이터들 중 'neutral or dissatisfied' 보다 'satisfied'의 수가 더 높은 Class는 어디 인가?

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/train.csv")

df_null = df.loc[(df["Arrival Delay in Minutes"].isnull())]
df_null.groupby(["Class","satisfaction"])["id"].count()

#Business
```

```python
#Answer
df.loc[df["Arrival Delay in Minutes"].isnull()].groupby(["Class", "satisfaction"],as_index = False).size().pivot(index='Class', columns = "satisfaction")
```

#작업형2

데이터 출처 :https://www.kaggle.com/teejmahal20/airline-passenger-satisfaction?select=train.csv (참고, 데이터 수정)

데이터 설명 : 비행탑승 경험 만족도 (satisfaction 컬럼 : 'neutral or dissatisfied' or satisfied ) 

문제타입 : 분류유형

평가지표 : accuracy

trainData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/train.csv

testData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/test.csv

```python
trainData = "https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/train.csv"
testData = "https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/test.csv"
subData = "https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/submission.csv"
import pandas as pd

train = pd.read_csv(trainData)
test = pd.read_csv(testData)
submission = pd.read_csv(subData)
train
```

```python
test
```

```python
#훈련데이터 NULL 체크
print(train.isnull().sum())
```

```python
#테스트 데이터 NULL 체크
test.isnull().sum()
```

```python
train.describe()
```

```python
#필요없는 컬럼 제거 및 타겟데이터 분리
#id, Arrival Delay in Minutes 

drop_col = ["id", "Arrival Delay in Minutes"]

train = train.drop(drop_col, axis=1)
test = test.drop(drop_col, axis=1)

X = train.drop("satisfaction", axis=1)
y = train["satisfaction"]
```

```python
#범주형 자료를 one-hot-encoding


columns = ["Gender", "Customer Type", "Class", "Type of Travel"]
X = pd.get_dummies(data=X, columns = columns)
test = pd.get_dummies(data=test, columns = columns)
```

```python
#스케일링(MinMaxScaling)

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X = scaler.fit_transform(X)
test = scaler.fit_transform(test)
```

```python
#홀드아웃
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X,y,test_size=0.3, stratify=y, random_state=42)
```

```python
#모델링
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)

rf.fit(X_train, y_train)

#평가
from sklearn.metrics import accuracy_score
y_pred = rf.predict(X_val)

print(accuracy_score(y_val, y_pred))
```

```python
pred = rf.predict(test)

submission["0"] = pred

#제출
submission.to_csv("수험번호", index = False)
```
