### Question 58

###Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라

###Unknown : N Less than 40K:a40K - 60K:b60K - 80K:c80K - 120K:d120K +’ : e

```python
mapping = {"Unknown" : "N",
           'Less than $40K': "a",
           '$40K - $60K' : "b",
           '$60K - $80K' : "c",
           '$80K - $120K' : "d",
           '$120K +' : "e"}

def changeCategory(x):
  if x == "Unknown":
    return "N"
  elif x == 'Less than $40K':
    return "a"
  elif x == '$40K - $60K':
    return "b"
  elif x == '$60K - $80K':
    return "c"
  elif x== '$80K - $120K':
    return "d"
  elif x=='$120K +':
    return "e"

df["newIncome"] = df["Income_Category"].apply(changeCategory)

df["newIncome"].head()
```

###Question 59

###Customer_Age의 값을 이용하여 나이 구간을 AgeState 컬럼으로 정의하라.

###(0 ~ 9 : 0 , 10 ~ 19 :10 , 20 ~ 29 :20 … 각 구간의 빈도수를 출력하라

```python
def changeState(x):
  for i in range(0,90,10):
    if i==0:
      if i <= x and x <= i+9:
        return i
    else:
      if x >= i and x <= i+9:
        return i


# def changeState(x):
#   if 0 <= x and x <= 9:
#     return 0
#   elif x >= 10 and x <= 19:
#     return 10
#   elif x >= 20 and x <= 29:
#     return 20
#   elif x >= 30 and x <= 39:
#     return 30
#   elif x >= 40 and x <= 49:
#     return 40
#   elif x >= 50 and x <= 59:
#     return 50
#   elif x >= 60 and x <= 69:
#     return 60
#   elif x >= 70 and x <= 79:
#     return 70

#df.Customer_Age.map(lamda x: x//10 *10)

df["AgeState"] = df["Customer_Age"].apply(changeState)

Ans = df["AgeState"].value_counts().sort_index()
Ans
```

###Question 62

###Marital_Status 컬럼값이 Married 이고 Card_Category 컬럼의 값이 Platinum인 경우 1 그외의 경우에는 모두 0으로 하는 newState컬럼을 정의하라. newState의 각 값들의 빈도수를 출력하라

```python
df["newState"] = 0

df.loc[(df["Marital_Status"] == "Married") & (df["Card_Category"] == "Platinum"), "newState"] = 1

def check(x):
  if x.Marital_Status == "Married" and x.Card_Category == "Platinum":
    return 1
  else:
    return 0

df["newState"] = df.apply(check, axis=1)

Ans = df["newState"].value_counts()

Ans
```
