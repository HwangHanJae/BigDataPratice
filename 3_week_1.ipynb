##Q1. price_range의 각 value를 그룹핑하여 각 그룹의 n_cores의 빈도가 가장 높은 value와 그 빈도수를 구하여라

```python
import pandas as pd 
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/mobile/train.csv")

df.groupby(["price_range"])["n_cores"].value_counts().sort_values(0).groupby(level=0).tail(1)

```

#작업형2

데이터 출처 : https://www.kaggle.com/iabhishekofficial/mobile-price-classification?select=train.csv

데이터 설명 : 핸드폰 가격예측 (price_range 칼럼 0(저렴) ~ 3(매우비쌈) 범위)

문제타입 : 분류유형
평가지표 : accuracy

trainData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/mobile/train.csv

testData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/mobile/test.csv

```python
trainData = "https://raw.githubusercontent.com/Datamanim/datarepo/main/mobile/train.csv"
testData = "https://raw.githubusercontent.com/Datamanim/datarepo/main/mobile/test.csv"

import pandas as pd

train  = pd.read_csv(trainData)
test = pd.read_csv(testData)

train
```

```python
test
```

```python
#타겟데이터 분리 price_range

X = train.drop(["price_range"], axis=1)
y = train["price_range"]

X
```

```python
#NULL 체크
print(X.isnull().sum())
#데이터 형태 확인
X.info()
```

```python
#학습데이터, 검증데이터 분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3, stratify = y)

#모델링
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()

rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

#평가
 
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, y_pred))
```

```python
#제출
#test 데이터에는 id값이 포함되어 있으니 id 제거

#test =test.drop("id", axis=1)
pred = rf.predict(test)

submission = {"price_range" : pred}
submission = pd.DataFrame(data=submission,)

submission.to_csv("수험번호.csv", index= False)
```
